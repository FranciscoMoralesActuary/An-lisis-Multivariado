# Escalonamiento Multidimensional


```python
from sklearn . manifold import MDS
import matplotlib . pyplot as plt
import pandas as pd
import numpy as np
```

```python
datos = {
"Estadistica": [10 , 9, 10, 8 , 9,7,10,5,10,8,6,10,10,9,10,9,9,7,7,6,9,7,8,8,6,8,9,9,9] ,
"Probabilidad": [10 , 9, 9 , 8 , 10,6,10,5,10,8,6,10,10,9,10,9,9,7,7,6,7,7,8,6,6,9,9,8,9] ,
"Ingles": [9 , 9, 10 , 6 , 6,10,10,5,9,8,6,5,10,9,8,9,9,7,7,6,9,7,8,7,5,9,9,9,10] ,
"Algebra": [8 , 8, 8 , 6 , 8,5,10,5,9,8,6,10,9,9,9,10,9,7,7,6,8,6,9,8,5,10,10,7,10] ,
"Demografia": [10 , 10, 10 , 10 , 10,10,10,5,10,8,6,10,10,9,10,10,9,7,7,6,8,6,10,8,5,8,10,8,9] ,
"Alumno": ["Francisco", "Diego", "Denilson", "Juan", "Tienda","Jesus","David","Yahir","Arturo","Hector",
           "Eduardo","Claudio","Paquito","Erick","Jessica","Rugerio","Luis","Ramon","Martin","Armenta","Ana",
           "Perla","Gabo","Esau","Adan","Apolinar","Marelyn","Oswaldo","Fernanda"]
 }
```

```python
df = pd.DataFrame(datos)
df
```

```python
X = df [[ "Estadistica", "Probabilidad","Ingles","Algebra","Demografia"]]
X
```

```python
mds = MDS(n_components = 2 , dissimilarity ="euclidean", random_state = 42)
datos_escalonados = mds.fit_transform(X)
```

```python
plt . figure ( figsize =(12 , 10) )
for clase in df ["Alumno"]. unique () :
  indices_clase = df ["Alumno"] == clase
  plt . scatter ( datos_escalonados [ indices_clase , 0] ,
    datos_escalonados [ indices_clase , 1], label = clase )


plt . title ("Escalamiento Multidimensional con Distancia Euclidiana")
plt . xlabel ("D i m e n s i o n 1")
plt . ylabel ("D i m e n s i o n 2")
plt . legend ()
plt . show ()
```

```python
distancias = {
    'Ciudad de México': [0, 537, 130, 2300, 997],
    'Guadalajara': [537, 0,  567, 2214, 1756],
    'Puebla': [130, 567,  0, 2342, 1003],
    'Tijuana': [2300, 2214, 2342, 0, 3943],
    'Mérida': [997, 1756,  1003, 3943, 0]
}
```

```python
df_distancias = pd.DataFrame(distancias, index=['Ciudad de México', 'Guadalajara', 'Puebla', 'Tijuana', 'Mérida'])
df_distancias
```

```python
mds = MDS(n_components=2, dissimilarity="precomputed", random_state=42)
datos_escalonados = mds.fit_transform(df_distancias)
```

```python
df_mds = pd.DataFrame(datos_escalonados, columns=['Dim1', 'Dim2'], index=df_distancias.index)
df_mds
```

```python

plt.figure(figsize=(10, 8))
plt.scatter(df_mds['Dim1'], df_mds['Dim2'])


for ciudad, (x, y) in df_mds.iterrows():
    plt.text(x, y, ciudad, fontsize=12, ha='right')


plt.title("Representación MDS de las distancias entre ciudades de México")
plt.xlabel("Dimensión 1")
plt.ylabel("Dimensión 2")
plt.grid()

plt.show()
```
